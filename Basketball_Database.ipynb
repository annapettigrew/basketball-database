```python
# Importing Dependencies 
import sqlalchemy
import pandas as pd
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# Creating an Engine 
engine = create_engine("sqlite:///basketball.sqlite")
conn = engine.connect()
```

```python
# Reading in the SQLITE File
player_attributes = pd.read_sql("SELECT * FROM player_attributes", conn)
player_salary = pd.read_sql("SELECT * FROM player_salary", conn)
draft = pd.read_sql("SELECT * FROM draft", conn)
draft
```

```python
# Merge the Columns in SQLite File
player_db = pd.merge(draft, player_salary, left_on="namePlayer", right_on="namePlayer", how="inner").merge(player_attributes, left_on="namePlayer", right_on="DISPLAY_FIRST_LAST", how="inner")
player_db = player_db.set_index('namePlayer')
player_db
```

```python
# Printing the Merged Column Names 
for col in player_db.columns:
    print(col)
```

```python
# Creating Column Recognization in SQLite File
new_player_db = player_db[['yearDraft','numberPickOverall','numberRound','numberRoundPick','slugTeam','nameOrganizationFrom','typeOrganizationFrom','idPlayer','idTeam','nameTeam_x','cityTeam','teamName','PLAYER_PROFILE_FLAG','slugOrganizationTypeFrom','locationOrganizationFrom','slugSeason','nameTeam_y','statusPlayer','isFinalSeason','isWaived','isOnRoster','isNonGuaranteed','isTeamOption','isPlayerOption','typeContractDetail','value','ID','FIRST_NAME','LAST_NAME','DISPLAY_FIRST_LAST','DISPLAY_LAST_COMMA_FIRST','DISPLAY_FI_LAST','PLAYER_SLUG','BIRTHDATE','SCHOOL','COUNTRY','LAST_AFFILIATION','HEIGHT','WEIGHT','SEASON_EXP','JERSEY','POSITION','ROSTERSTATUS','GAMES_PLAYED_CURRENT_SEASON_FLAG','TEAM_ID','TEAM_NAME','TEAM_ABBREVIATION','TEAM_CODE','TEAM_CITY','PLAYERCODE','FROM_YEAR','TO_YEAR','DLEAGUE_FLAG','NBA_FLAG','GAMES_PLAYED_FLAG','DRAFT_YEAR','DRAFT_ROUND','DRAFT_NUMBER','PTS','AST','REB','ALL_STAR_APPEARANCES','PIE']]
# Renaming Columns
new_player_db = new_player_db.rename(columns={
    'yearDraft':'DRAFT YEAR','namePlayer':'PLAYER NAME', 'numberPickOverall':'NUMBER OVERALL PICK', 'nameOrganizationFrom':'TEAM NAME FROM', 'typeOrganizationFrom':'ORGANIZATION FROM', 'nameTeam_x': 'TEAM NAME', 'slugSeason': 'SLUG SEASON', 'typeContractDetail':'TYPE OF CONTRACT', 'value':'SALARY', 'DISPLAY_FIRST_LAST':'NAME', 'JERSEY': 'JERSEY NUMBER'
})
# Dropping Columns
new_player_db = new_player_db.drop(['numberRoundPick', 'slugTeam', 'idPlayer', 'idTeam', 'teamName', 'PLAYER_PROFILE_FLAG','slugOrganizationTypeFrom','locationOrganizationFrom','nameTeam_y','statusPlayer', 'isFinalSeason', 'isWaived', 'isOnRoster', 'isNonGuaranteed','isTeamOption', 'isPlayerOption','NBA_FLAG','GAMES_PLAYED_FLAG','DRAFT_YEAR','DRAFT_NUMBER','PLAYERCODE','FROM_YEAR','TO_YEAR','DLEAGUE_FLAG','DRAFT_ROUND','ALL_STAR_APPEARANCES','GAMES_PLAYED_CURRENT_SEASON_FLAG','TEAM_ID','TEAM_NAME','TEAM_CODE','TEAM_CODE','TEAM_CITY','TEAM_ABBREVIATION','SEASON_EXP','ROSTERSTATUS','cityTeam','ID','FIRST_NAME','FIRST_NAME','LAST_NAME','COUNTRY','LAST_AFFILIATION','BIRTHDATE','SCHOOL','DISPLAY_FI_LAST','DISPLAY_LAST_COMMA_FIRST','PLAYER_SLUG','numberRound'], 1)
new_player_db
```

```python
# Importing CSV File
player_csv = ("Resources/NBA_PLAYERS_1996-2019.csv")
basketball_df = pd.read_csv(player_csv)
basketball_df
```

```python
# Printing Merged Column Names CSV File
for col in basketball_df.columns:
    print(col)
```

```python
# Creating Column Recognization 
new_basketball_df = basketball_df[['player_name','Unnamed: 0','team_abbreviation','age','player_height','player_weight','college','country','draft_year','draft_round','draft_number','gp','pts','reb','ast','net_rating','oreb_pct','dreb_pct','usg_pct','ts_pct','ast_pct','season']]

# Renaming Columns 
new_basketball_df = new_basketball_df.rename(columns={
    'gp':'GAMES PLAYED','player_name':'PLAYER NAME', 'pts': 'POINTS', 'reb':"REBOUNDS", 'net_rating':'NET RATING', 'ast':'ASSISTS','season': 'SEASON' 
})

# Dropping Columns 
new_basketball_df = new_basketball_df.drop(['Unnamed: 0', 'college', 'draft_round', 'draft_number', 'draft_year','age', 'player_height', 'player_weight','oreb_pct','dreb_pct','usg_pct','ts_pct','ast_pct','country','team_abbreviation'],1)
new_basketball_df
```

```python
# Merging CSV and SQLITE Files 
basketball_db = pd.merge(new_player_db, new_basketball_df, left_on="namePlayer", right_on="PLAYER NAME", how="inner")
basketball_db
```

```python
# Printing Merge Column Names in Merged Files 
for col in basketball_db.columns:
    print(col)
```

```python
# Creating Column Recognization
basketball_db = basketball_db[['PLAYER NAME','DRAFT YEAR','NUMBER OVERALL PICK','TEAM NAME FROM','ORGANIZATION FROM','TEAM NAME','SLUG SEASON','TYPE OF CONTRACT','SALARY','HEIGHT','WEIGHT','JERSEY NUMBER','POSITION','PTS','AST','REB','PIE','GAMES PLAYED','POINTS','REBOUNDS','ASSISTS','NET RATING','SEASON']]
basketball_db
```

```python
# Setting Index and Dropping Columns 
clean_basketball_db = basketball_db.set_index('PLAYER NAME')
clean_basketball_db = basketball_db.drop(['TYPE OF CONTRACT','SLUG SEASON'],1)
clean_basketball_db
```

```python
# Filtering by 2020-2021 Season 
season_db = clean_basketball_db[clean_basketball_db['SEASON'].str.contains("2020-21")]
season_db = season_db.drop_duplicates(subset='PLAYER NAME', keep="first")
season_db
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/new_basketball_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
clean_basketball_db.to_sql(name="basketball_db", con=engine, if_exists='append', index=False)
season_db.to_sql(name="season_db", con=engine, if_exists='append', index=False)
```
